```python
import tensorflow as tf
import numpy as np
```

# Define Weight 

```python
shape=[100,3,3,200]
variable=tf.Variable(tf.random_normal(shape , stddev=0.01))

```

# Start Session And Initialize Variable

```python
sess= tf.Session()
init= tf.initialize_all_variables()
sess.run(init)
```

# Save Weight to Numpy 

```python
var=sess.run(variable)
np_=np.asarray(var)
save_path='/home/seongjung/variable_save/'
np.save(save_path+'var' , var)


```

# Check Variable State

```python
print var[0:1]
```

# there is two way to restore Weight 

###### first one is by Using Numpy 


```python
np.save(save_path+'var' , var)
```

###### the other one is by Using Tensorflow Libraries 


```python
ckpt_dir="/home/seongjung/ckpt"
```

```python
saver=tf.train.Saver()
global_step = tf.Variable(0 , name='global_step' , trainable=False)
init= tf.initialize_all_variables()
sess.run(init)
```

```python
ckpt = tf.train.get_checkpoint_state(ckpt_dir)
print ckpt
```

```python
if ckpt and ckpt.model_checkpoint_path:
    print (ckpt.model_checkpoint_path)
    saver.restore(sess , ckpt.model_checkpoint_path)
```

```python
saver.save(sess, ckpt_dir+'/model.ckpt' , global_step = global_step)
```
