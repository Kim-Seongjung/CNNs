```python
import numpy as np 
import matplotlib.pyplot as plt
import os
```

```python
def extract_test_img(np_img ,img_row ,img_col , color_ch, crop_img_row , crop_img_col ):
    left_top =(0,0)
    right_top =(  img_row  - crop_img_row  , 0 )
    center =  ((img_row  - crop_img_row)/2  , (img_col - crop_img_row)/2)
    left_buttom = (0,(img_col - crop_img_row)/2 )
    right_buttom =  (img_row  - crop_img_row , img_col - crop_img_row)
    
    left_top_images  = np_img[: , left_top[0]:crop_img_row , left_top[1] : crop_img_col , :  ]
    right_top_images = np_img[: , right_top[0]:crop_img_row , right_top[1] : crop_img_col , :  ]
    center_images    = np_img[: , center[0]:crop_img_row , center[1] : crop_img_col , :  ]
    left_buttom_images=np_img[: , left_buttom[0]:crop_img_row , left_buttom[1] : crop_img_col , :  ]
    right_buttom_images= np_img[: , right_buttom[0]:crop_img_row , right_buttom[1] : crop_img_col , :  ]

    
        
    return left_top_images , right_top_images , center_images , left_buttom_images , right_buttom_images 

```

```python
os.getcwd()
img=np.load('/home/user01/notebook/eye_numpy_64/train_img.npy')
print np.shape(img)
np_img = img 
img_row = 128
img_col =128 
color_ch =3 
crop_img_row = 114
crop_img_col  =114 
left_top_images , right_top_images , center_images , left_buttom_images , right_buttom_images =extract_test_img(np_img ,img_row ,img_col , color_ch, crop_img_row , crop_img_col )

```

```python

print np.shape(left_top_images)

fig , axes = plt.subplots(2,2)
fig = plt.subplots_adjust(hspace = 0.3 , wspace =0.3 )

for i , ax in enumerate(axes.flat):
    ax.imshow(left_top_images[i])
plt.show()

```
